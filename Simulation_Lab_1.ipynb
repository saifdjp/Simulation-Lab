# **Witchman/Hill**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
R = list()
N = [100,1000,5000]
```

```python
for t in N:
    n=t
   
    z1=[12,7]
    u1=[]
    m1=16
    for i in range(2,n):
        new_z= (13*z1[i-1] + 11*z1[i-2] + 3)  % m1
        z1.append(new_z)
        u1_new = new_z/ m1
        u1.append(u1_new)

    #print(z1)
    #print(u1)

    z2= [3,5]
    u2= []
    m2=17
    for i in range(2,n):
        new_z2= (12*(z2[i-1]**2) + 13*z2[i-2])  %m2
        z2.append(new_z2)
        u2_new= new_z2 / m2
        u2.append(u2_new)
    #print(z2)
    #print(u2)

    z3 = [2,7]
    u3 = []
    m3 = 15
    for i in range(2,n):
        new_z3 = ( (z3[i-1]**3) + (z3[i-2]**2) ) % m3
        z3.append(new_z3)
        u3_new = new_z3 / m3
        u3.append(u3_new)


    u = []
    for i in range(0,n-2):
        u_new=u1[i]+ u2[i] +u3[i]
        u.append(u_new-int(u_new))

    print((u))  
    R.append(u)
```

```python
print("In 100 random, there are ",str(len(set(R[0]))), "number of unique random.")
print("In 1000 random, there are ",str(len(set(R[1]))), "number of unique random.")
print("In 5000 random, there are ",str(len(set(R[2]))), "number of unique random.")
```

```python
plt.figure(figsize=(60,20))
plt.bar(range(1000),u[:1000])
plt.xlabel("Instance")
plt.ylabel("Random Value")
```

```python
r = 3
q = 5
l = 4

b = [1,1,1,1,1]

N = [100]    #Number of random to generate
```

```python
for n in N:
    b = [1,1,1,1,1]
    no_of_bit = l*n                     # To get 'n' random we will need l*n bit
    
    for i in range(5,no_of_bit):
        new_b = (b[i-r]+b[i-q]) % 2
        b.append(new_b)
    z = list()
    for i in range(3,len(b),4):
        new_z = b[i-3:i+1]
        new_z= sum(np.array(new_z)*[8,4,2,1])
        z.append(new_z)
    print(z)
    
    u = np.array(z)/16      
    plt.figure(figsize=(150,40))
    plt.bar(range(n),u)
```
